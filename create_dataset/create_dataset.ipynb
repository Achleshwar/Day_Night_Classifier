```python
import os
import sys
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
training_data = []
def create_training_data():
    DATADIR = 'C:/Users/Achleshwar Luthra/Desktop/Data-Science/Day-Night Classifier/Day_images'
    class_name = 0
    for img in os.listdir(DATADIR):
        try:
            img_array = cv2.imread(os.path.join(DATADIR, img))
            img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
            training_data.append([img_array, class_name])
        except Exception as exc:
            pass
        
create_training_data()
```

```python
print(len(training_data))
```

```python
#neeed this after both day and night images
import random 
random.shuffle(training_data)

for sample in training_data[:10]:
    print(sample[1])
```

```python
X =[]
y = []
```

```python
IMG_SIZE = 64
for features, label in training_data:
    img_array = cv2.resize(features, (IMG_SIZE, IMG_SIZE))
    #append them to X and y
    X.append(img_array)
    y.append(label)
    
    
X = np.array(X).reshape(-1,IMG_SIZE, IMG_SIZE,3)
```

```python
import pickle

pickle_out = open("Day_X.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("Day_y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```
